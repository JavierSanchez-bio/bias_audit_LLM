```python
import json
import os
from unsloth import FastLanguageModel

ruta_config = "first_bias/adapter_config.json"
with open(ruta_config, "r") as f:
    config_data = json.load(f)

config_data["base_model_name_or_path"] = "unsloth/llama-3-8b-bnb-4bit"

with open(ruta_config, "w") as f:
    json.dump(config_data, f, indent=4)


model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "first_bias",
    max_seq_length = 2048,
    load_in_4bit = True,
)

model.save_pretrained_merged("modelo_fusionado", tokenizer, save_method = "merged_16bit")

if os.path.exists("modelo_fusionado"):
    print("Success, modelo fusionado")
else:
    print("Error: La carpeta no se creó.")
```

```python
print("Iniciando creación del GGUF:")

model.save_pretrained_gguf(
    "modelo_final",
    tokenizer,
    quantization_method = "q4_k_m"
)

print("Proceso terminado")
```
